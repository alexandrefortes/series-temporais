**About Dataset**  
Some say climate change is the biggest threat of our age while others say it‚Äôs a myth based on dodgy science. We are turning some of the data over to you so you can form your own view.

Even more than with other data sets that Kaggle has featured, there‚Äôs a huge amount of data cleaning and preparation that goes into putting together a long-time study of climate trends. Early data was collected by technicians using mercury thermometers, where any variation in the visit time impacted measurements. In the 1940s, the construction of airports caused many weather stations to be moved. In the 1980s, there was a move to electronic thermometers that are said to have a cooling bias.

Given this complexity, there are a range of organizations that collate climate trends data. The three most cited land and ocean temperature data sets are NOAA‚Äôs MLOST, NASA‚Äôs GISTEMP and the UK‚Äôs HadCrut.

We have repackaged the data from a newer compilation put together by the Berkeley Earth, which is affiliated with Lawrence Berkeley National Laboratory. The Berkeley Earth Surface Temperature Study combines 1.6 billion temperature reports from 16 pre-existing archives. It is nicely packaged and allows for slicing into interesting subsets (for example by country). They publish the source data and the code for the transformations they applied. They also use methods that allow weather observations from shorter time series to be included, meaning fewer observations need to be thrown away.

In this dataset, we have include several files:

Global Land and Ocean-and-Land Temperatures (GlobalTemperatures.csv):

Date: starts in 1750 for average land temperature and 1850 for max and min land temperatures and global ocean and land temperatures

- LandAverageTemperature: global average land temperature in celsius  
- LandAverageTemperatureUncertainty: the 95% confidence interval around the average  
- LandMaxTemperature: global average maximum land temperature in celsius  
- LandMaxTemperatureUncertainty: the 95% confidence interval around the maximum land temperature  
- LandMinTemperature: global average minimum land temperature in celsius  
- LandMinTemperatureUncertainty: the 95% confidence interval around the minimum land temperature  
- LandAndOceanAverageTemperature: global average land and ocean temperature in celsius  
- LandAndOceanAverageTemperatureUncertainty: the 95% confidence interval around the global average land and ocean temperature  

Other files include:  

- Global Average Land Temperature by Country (GlobalLandTemperaturesByCountry.csv) 
- Global Average Land Temperature by State (GlobalLandTemperaturesByState.csv) 
- Global Land Temperatures By Major City (GlobalLandTemperaturesByMajorCity.csv) 
- Global Land Temperatures By City (GlobalLandTemperaturesByCity.csv) 

The raw data comes from the [Berkeley Earth](http://berkeleyearth.org/about/) data page.

## Avalia√ß√£o da s√©rie

```python
# Fun√ß√£o para an√°lise de uma s√©rie temporal
def analyze_time_series(series):
    # DataFrame para armazenar os resultados
    results = pd.DataFrame(columns=["Metric", "Value"])

    # Estacionariedade - Teste Augmented Dickey-Fuller
    adf_test = adfuller(series, autolag='AIC')
    results = results.append({"Metric": "ADF Statistic", "Value": adf_test[0]}, ignore_index=True)
    results = results.append({"Metric": "ADF p-value", "Value": adf_test[1]}, ignore_index=True)

    # Tend√™ncia e Sazonalidade - Decomposi√ß√£o
    decomposition = seasonal_decompose(series, model='additive', period=12) # Assumindo uma sazonalidade mensal
    trend = decomposition.trend.dropna()
    seasonal = decomposition.seasonal.dropna()

    # Verificando se existe tend√™ncia
    has_trend = not np.all(np.isnan(trend))
    results = results.append({"Metric": "Has Trend", "Value": has_trend}, ignore_index=True)

    # Verificando se existe sazonalidade
    has_seasonality = not np.all(np.isnan(seasonal))
    results = results.append({"Metric": "Has Seasonality", "Value": has_seasonality}, ignore_index=True)

    # Autocorrela√ß√£o
    autocorr = acf(series)
    results = results.append({"Metric": "Autocorrelation (lag 1)", "Value": autocorr[1]}, ignore_index=True)

    # Autocorrela√ß√£o Parcial
    pac = pacf(series)
    results = results.append({"Metric": "Partial Autocorrelation (lag 1)", "Value": pac[1]}, ignore_index=True)

    return results
```

```python
# Exemplo de uso da fun√ß√£o com dados fict√≠cios
np.random.seed(42)
your_series = pd.Series(np.random.randn(100))
results = analyze_time_series(your_series)
results
```

### Consulta API OpenAI

```python
from dotenv import load_dotenv
import os
from openai import OpenAI

def fala_gepeto(prompt, temperature=0.7, model="gpt-3.5", max_tokens=300, response_format="json"):

    # Load environment variables from .env file
    load_dotenv()
    api_key=os.getenv("OPENAI_API_KEY")
    openai_client = OpenAI(api_key=api_key)

    # Determinar o modelo com base nos par√¢metros
    if model == "gpt-4":
        model_to_use = "gpt-4-1106-preview"
    else:
        model_to_use = "gpt-3.5-turbo"

    # Configura√ß√£o da resposta
    if response_format == "json" and model == "gpt-4":
        response_config={"type": "json_object"}
        prompt+="Resposta em json."
    else:
        response_config=None  

    # Criando o prompt para a API
    completion = openai_client.chat.completions.create(
        model=model_to_use,
        response_format=response_config,
        temperature=temperature,
        max_tokens=max_tokens,
        messages=[
            {"role": "system", "content": "Voc√™ √© um Analista de S√©ries Temporais especialista em clima do planeta Terra."},
            {"role": "user", "content": prompt}
        ]
    )
    return completion.choices[0].message.content
```

### Carregar datasets

```python
def carregar_dataset_por_nome(nome, coluna_data, colunas_interesse=None, filtro=None):
    datasets = {
        "Country": 'earth-temperature/GlobalLandTemperaturesByCountry.csv',
        "State": 'earth-temperature/GlobalLandTemperaturesByState.csv',
        "MajorCity": 'earth-temperature/GlobalLandTemperaturesByMajorCity.csv',
        "City": 'earth-temperature/GlobalLandTemperaturesByCity.csv'
    }

    if nome not in datasets:
        print("Nome do dataset inv√°lido")
        return None

    path = datasets[nome]
    print(f"Carregando dataset: {nome}")
    df = pd.read_csv(path)

    # Filtrar por valor espec√≠fico em uma coluna
    if filtro:
        for coluna, valor in filtro.items():
            df = df[df[coluna] == valor]

    # Remover linhas com valores nulos ou brancos
    df = df.dropna().replace('', pd.NA).dropna()
    
    # Filtrar colunas de interesse
    if colunas_interesse:
        df = df[colunas_interesse]

    df[coluna_data] = pd.to_datetime(df[coluna_data])
    df.set_index(coluna_data, inplace=True)
    # Ordena os dados por data
    df=df.sort_values(by=coluna_data)

    return df
```

### Preparo de Dados para LSTM

```python
def preparar_dados_lstm(df, coluna_objetivo, janela_temporal=1):
    df = df.fillna(method='ffill').fillna(method='bfill')
    
    # Normalizar dados
    scaler = MinMaxScaler(feature_range=(0, 1))
    df_scaled = scaler.fit_transform(df)

    X, y = [], []
    for i in range(len(df_scaled) - janela_temporal - 1):
        X.append(df_scaled[i:(i + janela_temporal), :])
        y.append(df_scaled[i + janela_temporal, df.columns.get_loc(coluna_objetivo)])
    
    return np.array(X), np.array(y)
```

```python
# Exemplo de DataFrame
country_df = carregar_dataset_por_nome("Country", "dt",
                                       ["dt", "AverageTemperature"], 
                                       {"Country": "Brazil"})

country_df.head()                                       

# Uso da fun√ß√£o
country_X, country_y = preparar_dados_lstm(country_df, 'AverageTemperature', janela_temporal=2)
# country_X, country_y s√£o agora arrays do numpy prontos para serem usados em um modelo LSTM
```

### Preparo de Dados para T√©cnicas de S√©ries Temporais

```python
def preparar_dados_series_temporais(df, coluna_objetivo, num_lags=1):
    df = df.fillna(method='ffill').fillna(method='bfill')
    
    # Criar lags (atrasos) da s√©rie temporal
    for lag in range(1, num_lags + 1):
        df[f'lag_{lag}'] = df[coluna_objetivo].shift(lag)

    # Remover linhas com valores nulos criados pelos lags
    df = df.dropna()

    # Separar em vari√°veis explicativas (X) e vari√°vel alvo (y)
    X = df[[col for col in df.columns if col.startswith('lag_')]]
    y = df[coluna_objetivo]

    return X, y
```

```python
# Exemplo de DataFrame
country_df = carregar_dataset_por_nome("Country", "dt",
                                       ["dt", "AverageTemperature"], 
                                       {"Country": "Brazil"})

# Uso da fun√ß√£o
country_X, country_y = preparar_dados_series_temporais(country_df, 'AverageTemperature', num_lags=2)
print(country_X, country_y)
# country_X, country_y s√£o agora arrays do numpy prontos para serem usados em um modelo LSTM
```

### Registro de Performance

```python
import pandas as pd

def criar_registro_modelos():
    colunas = ['modelo', 'metricas_desempenho', 'hardware_treinamento', 'hyperparametros', 'tempo_processamento']
    return pd.DataFrame(columns=colunas)

def adicionar_registro_performance(df, modelo, metricas_desempenho, hardware_treinamento, hyperparametros, tempo_processamento):
    novo_registro = {
        "modelo": modelo,
        "metricas_desempenho": metricas_desempenho,
        "hardware_treinamento": hardware_treinamento,
        "hyperparametros": hyperparametros,
        "tempo_processamento": tempo_processamento
    }
    # Atualiza se o modelo j√° existir, sen√£o adiciona um novo registro
    if modelo in df['modelo'].values:
        df.loc[df['modelo'] == modelo] = novo_registro
    else:
        df = pd.concat([df, pd.DataFrame([novo_registro])], ignore_index=True)
    return df
```

```python
# Exemplo de uso
performances_df = criar_registro_modelos()
performances_df = adicionar_registro_performance(performances_df, "LSTM", 
                                                 {"MAE": 0.5, "RMSE": 0.7, "MAPE": 0.2, "R¬≤": 0.85},
                                                 "GPU NVIDIA RTX 3080", 
                                                 {"batch_size": 32, "epochs": 100, "learning_rate": 0.01},
                                                 3600)  # Tempo de processamento em segundos

# Convertendo para JSON
json_performances_df = performances_df.to_json(orient="records")
print(json_performances_df)
```

### ARIMA (Autoregressive Integrated Moving Average)

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt

country_df = carregar_dataset_por_nome("Country", "dt", ["dt", "AverageTemperature"], {"Country": "Brazil"})

# Fun√ß√£o para realizar o teste ADF e imprimir os resultados
def test_stationarity(timeseries):
    # Estat√≠sticas de rolamento
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()

    # Plotar estat√≠sticas de rolamento
    plt.figure(figsize=(12, 6))
    plt.plot(timeseries, color='blue', label='Original')
    plt.plot(rolling_mean, color='red', label='M√©dia M√≥vel')
    plt.plot(rolling_std, color='black', label='Desvio Padr√£o M√≥vel')
    plt.legend(loc='best')
    plt.title('Estat√≠sticas de Rolamento')
    plt.show()

    # Realizar o teste ADF
    print('Resultados do Teste Dickey-Fuller:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Estat√≠stica do Teste', 'Valor-p', 'Lags Usados', 'N√∫mero de Observa√ß√µes Usadas'])
    for key, value in dftest[4].items():
        dfoutput['Valor Cr√≠tico (%s)' % key] = value
    print(dfoutput)

# Chame a fun√ß√£o de teste de estacionariedade
test_stationarity(country_df['AverageTemperature'])
```

Os resultados do Teste Dickey-Fuller s√£o usados para determinar se uma s√©rie temporal √© estacion√°ria. Aqui est√° a interpreta√ß√£o dos principais resultados do teste:

1. **Estat√≠stica do Teste (-3.012049)**: Este valor √© a estat√≠stica do teste ADF. Quanto mais negativo for, maior a evid√™ncia de que a s√©rie √© estacion√°ria. Neste caso, a estat√≠stica do teste √© -3.012049, o que sugere uma forte evid√™ncia de estacionariedade.

2. **Valor-p (0.033784)**: O valor-p √© a probabilidade associada √† estat√≠stica do teste. Ele indica a probabilidade de obter uma estat√≠stica de teste t√£o extrema quanto a observada, se a hip√≥tese nula de n√£o estacionariedade for verdadeira. Neste caso, o valor-p √© 0.033784, que √© menor que o n√≠vel de signific√¢ncia comum de 0,05. Isso significa que voc√™ pode rejeitar a hip√≥tese nula de n√£o estacionariedade e concluir que a s√©rie √© estacion√°ria.

3. **Lags Usados (26.000000)**: Este √© o n√∫mero de lags (atrasos) inclu√≠dos no teste. Em geral, um n√∫mero maior de lags pode fornecer uma estimativa mais precisa, mas tamb√©m torna o teste menos sens√≠vel. Neste caso, foram usados 26 lags.

4. **N√∫mero de Observa√ß√µes Usadas (2137.000000)**: Este √© o n√∫mero de observa√ß√µes da s√©rie temporal que foram usadas no teste. Neste caso, 2137 observa√ß√µes foram usadas.

5. **Valores Cr√≠ticos (1%, 5%, 10%)**: Estes s√£o os valores cr√≠ticos associados aos n√≠veis de signific√¢ncia de 1%, 5% e 10%. Eles s√£o usados para comparar com a estat√≠stica do teste. Se a estat√≠stica do teste for menor do que esses valores cr√≠ticos, voc√™ pode rejeitar a hip√≥tese nula de n√£o estacionariedade. Neste caso, a estat√≠stica do teste (-3.012049) √© maior do que o valor cr√≠tico de 1% (-3.433414), mas menor do que os valores cr√≠ticos de 5% (-2.862893) e 10% (-2.567491). Isso significa que voc√™ pode rejeitar a hip√≥tese nula com um n√≠vel de signific√¢ncia de 5% ou 10%, mas n√£o com um n√≠vel de signific√¢ncia de 1%.

Resumindo, com base nos resultados do Teste Dickey-Fuller, a s√©rie temporal parece ser estacion√°ria, pois o valor-p √© menor que 0,05 (n√≠vel de signific√¢ncia comum) e a estat√≠stica do teste sugere evid√™ncia de estacionariedade. No entanto, voc√™ pode querer considerar outros fatores e realizar mais an√°lises antes de concluir definitivamente sobre a estacionariedade da s√©rie.

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import time
import matplotlib.pyplot as plt

country_df = carregar_dataset_por_nome("Country", "dt", ["dt", "AverageTemperature"], {"Country": "Brazil"})

# Converter √≠ndice para datetime se ainda n√£o estiver
country_df.index = pd.to_datetime(country_df.index)

# Tratando valores ausentes
# Op√ß√£o 1: Preenchimento para frente (forward fill)
country_df['AverageTemperature'].fillna(method='ffill', inplace=True)

# Separar os dados em conjunto de treinamento e teste
cutoff = int(len(country_df) * 0.8)
train_data = country_df.iloc[:cutoff]
test_data = country_df.iloc[cutoff:]

# Ajuste do modelo ARIMA
start_time = time.time()
model = ARIMA(train_data['AverageTemperature'], order=(5,1,0))  # Exemplo de hiperpar√¢metros p=5, d=1, q=0
fitted_model = model.fit()
end_time = time.time()

# Previs√µes
predictions = fitted_model.forecast(steps=len(test_data))
predictions = pd.Series(predictions, index=test_data.index)

# Verificando a presen√ßa de NaNs
print("NaNs em test_data:", test_data['AverageTemperature'].isna().sum())
print("NaNs em predictions:", predictions.isna().sum())

# Tratamento de NaNs em test_data
if test_data['AverageTemperature'].isna().sum() > 0:
    test_data['AverageTemperature'].fillna(method='ffill', inplace=True)  # Forward fill
    # Ou, use interpola√ß√£o
    # test_data['AverageTemperature'].interpolate(method='linear', inplace=True)

# Tratamento de NaNs em predictions
if predictions.isna().sum() > 0:
    #predictions.fillna(method='ffill', inplace=True)  # Forward fill
    # Ou, use o √∫ltimo valor conhecido
    predictions.fillna(test_data['AverageTemperature'].iloc[-1], inplace=True)

# Recalcular as m√©tricas
mae = mean_absolute_error(test_data['AverageTemperature'], predictions)
rmse = np.sqrt(mean_squared_error(test_data['AverageTemperature'], predictions))
mape = np.mean(np.abs(predictions - test_data['AverageTemperature']) / np.abs(test_data['AverageTemperature']))
r2 = r2_score(test_data['AverageTemperature'], predictions)

# Plotar a performance
plt.figure(figsize=(10, 6))
plt.plot(train_data['AverageTemperature'], label='Treinamento')
plt.plot(test_data['AverageTemperature'], label='Teste')
plt.plot(predictions, label='Previs√µes ARIMA')
plt.title('Performance do Modelo ARIMA')
plt.xlabel('Data')
plt.ylabel('Temperatura M√©dia')
plt.legend()
plt.show()

# Adicionar registro de performance
performances_df = adicionar_registro_performance(
    performances_df, "ARIMA", {"MAE": mae, "RMSE": rmse, "MAPE": mape, "R¬≤": r2},
    "CPU", {"p": 5, "d": 1, "q": 0}, end_time - start_time
)
```

### LSTM (Long Short-Term Memory networks)

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import time
import matplotlib.pyplot as plt

country_df = carregar_dataset_por_nome("Country", "dt", ["dt", "AverageTemperature"], {"Country": "Brazil"})

# Fun√ß√£o para converter s√©rie para aprendizado supervisionado
def series_to_supervised(data, n_in=1, n_out=1):
    df = pd.DataFrame(data)
    cols = list()
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
    for i in range(0, n_out):
        cols.append(df.shift(-i))
    agg = pd.concat(cols, axis=1)
    agg.dropna(inplace=True)
    return agg.values

# Preparando os dados
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(country_df[['AverageTemperature']])

# Par√¢metros do modelo
n_input = 5
n_features = 1
batch_size = 32
epochs = 100
neurons = 50
dropout = 0.2
layers = 2

# Convertendo a s√©rie temporal para formato supervisionado
supervised = series_to_supervised(scaled_data, n_input)
X, y = supervised[:, :-1], supervised[:, -1]

# Divis√£o em conjuntos de treinamento e teste
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Redimensionando entrada para [amostras, etapas de tempo, caracter√≠sticas]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], n_features))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], n_features))

# Construindo o modelo LSTM
model = Sequential()
model.add(LSTM(neurons, activation='relu', input_shape=(n_input, n_features), return_sequences=(layers > 1)))
for i in range(layers - 1):
    model.add(LSTM(neurons, activation='relu', return_sequences=(i < layers - 2)))
    model.add(Dropout(dropout))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Early Stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

# Treinamento do modelo
start_time = time.time()
history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test), verbose=2, callbacks=[early_stopping])
end_time = time.time()

# Previs√µes
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

# C√°lculo das m√©tricas de performance
mae = mean_absolute_error(y_test_inv, predictions)
rmse = np.sqrt(mean_squared_error(y_test_inv, predictions))
mape = np.mean(np.abs(predictions - y_test_inv) / np.abs(y_test_inv))

# Plotando a performance
plt.figure(figsize=(10, 6))
plt.plot(y_test_inv, label='Real')
plt.plot(predictions, label='Previs√µes LSTM')
plt.title('Performance do Modelo LSTM')
plt.xlabel('Tempo')
plt.ylabel('Temperatura M√©dia')
plt.legend()
plt.show()

# Adicionar registro de performance
performances_df = adicionar_registro_performance(
    performances_df, "LSTM", {"MAE": mae, "RMSE": rmse, "MAPE": mape},
    "GPU NVIDIA RTX 4070 TI", {"neurons": neurons, "epochs": epochs, "batch_size": batch_size, "layers": layers, "dropout": dropout},
    end_time - start_time
)
```
